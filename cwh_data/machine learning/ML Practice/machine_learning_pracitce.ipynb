```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_excel("iris_dataset.xlsx")
```

```python
data_train = pd.read_excel('iris-train.xlsx')
data_test = pd.read_excel('ris-test.xlsx')
```

```python
data_train.head()
```

```python
print(data_train.shape)
print(data_test.shape)
```

```python
x = data_train.iloc[:,:-1] #take all rows (:) and all columsn expect the last column (:-1)
```

```python
y = data_train.iloc[:,-1] #take all rows (:) and only the last column (-1)
```

```python
print(x)
```

```python
print(y)
```

```python
model  = RandomForestClassifier()
```

```python
model.fit(x,y)
```

```python
predict = model.predict([[5.4,3,4.5,1.5]])
```

```python
predict
```

```python
x_test = data_test.iloc[:,:-1]
y_test = data_test.iloc[:,-1]
```

```python
x_test.values
```

```python
predict_whole_test_data = model.predict(x_test.values)
```

```python
predict_whole_test_data
```

```python
y_test
```

```python
y_test.values
```

```python
y_test == predict_whole_test_data
```

```python
actual_values = y_test.values
matched_count = 0
unmatched_count = 0
count = 0

for i in range(len(actual_values)):
    count+=1
    if actual_values[i] == predict_whole_test_data[i]:
        print("value_matched")
        matched_count += 1
    else:
        print("unmatched value")
        unmatched_count += 1

print("Matched count:", matched_count)
print("Unmatched count:", unmatched_count)
accuracy = round((matched_count/count)*100,2)

print("accuracy", accuracy,"%")
```
