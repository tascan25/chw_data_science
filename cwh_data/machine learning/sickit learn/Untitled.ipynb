```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Features: [weight (grams), color (0 = red, 1 = orange)]
features = [
    [150, 0],  # Apple
    [170, 0],  # Apple
    [130, 1],  # Orange
    [120, 1],  # Orange
]
labels = ["apple", "apple", "orange", "orange"]
```

```python
classifier = DecisionTreeClassifier()
```

```python
classifier.fit(features,labels)
```

```python
print(classifier.predict([[160, 0]]))  # Heavy and red-ish? Probably an apple!
print(classifier.predict([[115, 1]]))  # Light and orange? Probably an orange!
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt 
```

```python
plt.figure(figsize=(160,6))
plot_tree(classifier,feature_names=["weight","color"], class_names=["red","orange"],filled=True)
plt.show()
```
