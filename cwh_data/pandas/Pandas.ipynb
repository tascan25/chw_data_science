## MELT AND PIVOT

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}
 
df = pd.DataFrame(data)
 
# Display the DataFrame
print(df)
```

```python



'''
id_vars, will take a column that we want to keep as it is..
value_vars, will take a columns name that we want to melt
var_name, will take a name of the column that we want to assign to the column in which we have melt the value
value_name, will take a nmaem which we want to give it, which will contain the values of the columns that we have melt
'''
```

```python
df.melt(id_vars=["Name"],value_vars=["Math","Science","English"],var_name="Subject",value_name="Score")
```

```python
df.pivot(index="Name", columns="Subject", values="Score")
```

## AGGREGATION AND GROUPING

```python
import pandas as pd

df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})  

```

```python
df
```

```python
df.groupby('Department')['Salary'].mean()
```

```python
df.groupby('Team')['Salary'].sum()
```

```python
df.groupby('Department')['Team'].count()
```

```python
df.groupby('Team')['Salary'].agg(['mean','max','min'])
```

```python
type(df.groupby('Team')['Salary'].agg(['mean','max','min']))
```

```python
df.groupby('Team')['Salary'].aggregate(['mean','max','min'])
```

## MERGING AND JOINING DATA

```python
import pandas as pd

employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})
 
departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
```

```python
employees
```

```python
departments
```

```python
pd.merge(employees,departments,on="DeptID") ##inner join
```

```python
pd.merge(employees,departments,on="DeptID",how="left") #left join
```

```python
pd.merge(employees,departments,on="DeptID",how="right") #left join
```

```python
pd.merge(employees,departments,on="DeptID",how="outer") #left join
```

```python
df1 = pd.DataFrame({"Name": ["Alice", "Bob"]})
df2 = pd.DataFrame({"Name": ["Charlie", "David"]})
```

```python
 
pd.concat([df1, df2],axis=1)
```
