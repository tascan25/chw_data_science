# REQUEST MODULE FOR DATA COLLECTION

```python
import requests
```

```python
a = requests.get('https://books.toscrape.com/')
```

```python
with open("htmls/page1.html","w") as f:
    f.write(a.text)
```

```python
for i in range(1,51):
    a = requests.get(f'https://books.toscrape.com/catalogue/page-{i}.html')
    with open(f"htmls/page{i}.html","w",encoding="utf-8") as f:
        f.write(a.text)
        print(f"downloading page {i} successfully!!")
```

# BEAUTIFUL SOUP

```python
from bs4 import BeautifulSoup
```

```python
with open('htmls/page1.html') as f:
    content = f.read()
```

```python
soup = BeautifulSoup(content,"html.parser")
```

```python
h3s = soup.find_all("h3")
```

```python
h3s
```

```python
for h3 in h3s:
    print(h3.find("a")['title'])
```

```python
all_page1_articles = soup.select("article.product_pod")
```

```python
all_page1_articles
```

```python
items_page1 = []
for article in all_page1_articles:
    title = article.find("h3").find("a")['title']
    price = article.select_one("p.price_color").text.split("£")[1]
    rating_element = article.select_one("p.star-rating")
    rating = rating_element['class'][1]
    items_page1.append([title,price,rating])
items_page1
```

```python
import pandas as pd
```

```python
df_page_1 = pd.DataFrame(items_page1, columns=['Book Title', "Price", "Rating"])
```

```python
df_page_1
```

## for all 50 pages

```python
content = []
for i in range(1,51):
    with open(f'htmls/page{i}.html') as f:   
        content.append(f.read())
combined_html = "".join(content)
# bigger_soup = BeautifulSoup(content,"html.parser")

```

```python


bigger_soup = BeautifulSoup(combined_html,"html.parser")
```

```python
all_pages_article = bigger_soup.select("article.product_pod")
```

```python
items_all_pages = []
for article in all_pages_article:
    title = article.find("h3").find("a")['title']
    price = article.select_one("p.price_color").text.split("£")[1]
    rating_element = article.select_one("p.star-rating")
    rating = rating_element['class'][1]
    items_all_pages.append([title,price,rating])
items_all_pages
```

```python
df_all_page = pd.DataFrame(items_all_pages, columns=["Book Title", "Price", "Rating"])
```
